```python
import pickle

with open("repo_info_train_doc_embeddings.pkl", "rb") as f:
    repo_info_train_doc_embeddings = pickle.load(f)
    f.close()
```

```python
import torch
device = "cpu"
```

```python
for repo_name, repo_dict in repo_info_train_doc_embeddings.items():
    docs_embeddings = repo_info_train_doc_embeddings[repo_name]["docs_embeddings"]
    if type(docs_embeddings) is torch.Tensor:
        repo_info_train_doc_embeddings[repo_name]["docs_embeddings"] = repo_info_train_doc_embeddings[repo_name]["docs_embeddings"].to(device)
        print(repo_info_train_doc_embeddings[repo_name]["docs_embeddings"].get_device())
    else:
        for model_name, model_embedding in docs_embeddings.items():
            repo_info_train_doc_embeddings[repo_name]["docs_embeddings"][model_name] = repo_info_train_doc_embeddings[repo_name]["docs_embeddings"][model_name].to(device)
            print(repo_info_train_doc_embeddings[repo_name]["docs_embeddings"][model_name].get_device())
```

```python
with open("repo_info_train_doc_embeddings.pkl", "wb") as f:
    pickle.dump(repo_info_train_doc_embeddings, f)
    f.close()
```
